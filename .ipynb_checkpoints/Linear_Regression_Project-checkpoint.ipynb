```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
import os
print(os.getcwd())
```

```python
data = pd.read_csv('Salary_dataset.csv')  # Just the filename
print(data.head())
```

```python
# Extract X and Y columns
X = data['YearsExperience'].values.reshape(-1, 1)  # Reshape for sklearn
Y = data['Salary'].values

print(f"Number of samples: {len(X)}")
```

```python
# Create linear regression model
model = LinearRegression()

# Train the model
model.fit(X, Y)

# Get the coefficients
slope = model.coef_[0]
intercept = model.intercept_

print(f"Regression equation: Y = {slope:.2f}X + {intercept:.2f}")
```

```python
# Make predictions
Y_pred = model.predict(X)

# Calculate metrics
r2 = r2_score(Y, Y_pred)
mse = mean_squared_error(Y, Y_pred)
rmse = np.sqrt(mse)

print(f"\nModel Performance:")
print(f"R² Score: {r2:.4f}")
print(f"RMSE: {rmse:.2f}")
```

```python
# Create scatter plot with regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Actual Data', alpha=0.6)
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience - Simple Linear Regression')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import pandas as pd

# Load data
phone_data = pd.read_csv('phone_usage.csv', sep=';')
phone_data.columns = ['ST', 'MRC', 'DU']

# Extract variables
n = len(phone_data)
X1 = phone_data['ST'].values
X2 = phone_data['MRC'].values
Y = phone_data['DU'].values

# Calculate all the sums from your table
sum_X1 = np.sum(X1)
sum_X2 = np.sum(X2)
sum_Y = np.sum(Y)
sum_X1_squared = np.sum(X1**2)
sum_X2_squared = np.sum(X2**2)
sum_Y_squared = np.sum(Y**2)
sum_X1X2 = np.sum(X1 * X2)
sum_X1Y = np.sum(X1 * Y)
sum_X2Y = np.sum(X2 * Y)

print("Verification of sums from your table:")
print(f"n = {n}")
print(f"Σx₁ = {sum_X1:.1f}")
print(f"Σx₂ = {sum_X2:.0f}")
print(f"Σy = {sum_Y:.1f}")
print(f"Σx₁² = {sum_X1_squared:.2f}")
print(f"Σx₂² = {sum_X2_squared:.0f}")
print(f"Σy² = {sum_Y_squared:.2f}")
print(f"Σx₁x₂ = {sum_X1X2:.1f}")
print(f"Σx₁y = {sum_X1Y:.2f}")
print(f"Σx₂y = {sum_X2Y:.1f}")
```

```python
# Calculate b1 using the formula from your report
# The normal equations for multiple regression
numerator_b1 = (sum_X2_squared * sum_X1Y) - (sum_X1X2 * sum_X2Y)
denominator = (sum_X1_squared * sum_X2_squared) - (sum_X1X2**2)
b1 = numerator_b1 / denominator

# Calculate b2
numerator_b2 = (sum_X1_squared * sum_X2Y) - (sum_X1X2 * sum_X1Y)
b2 = numerator_b2 / denominator

# Calculate a (intercept)
a = (sum_Y - (b1 * sum_X1) - (b2 * sum_X2)) / n

print("\nRegression Coefficients (Manual Calculation):")
print(f"a (intercept) = {a:.2f}")
print(f"b₁ (ST coefficient) = {b1:.4f}")
print(f"b₂ (MRC coefficient) = {b2:.5f}")

print(f"\nRegression Equation:")
print(f"Y = {a:.2f} + {b1:.4f}X₁ + {b2:.5f}X₂")
```

```python
# Calculate R² using the formula from your report
# R² = (b₁Σx₁y + b₂Σx₂y) / Σy²

R2_manual = (b1 * sum_X1Y + b2 * sum_X2Y) / sum_Y_squared

print(f"\nCoefficient of Determination:")
print(f"R² = {R2_manual:.2f}")
print(f"R² = {R2_manual:.4f} (more precision)")

# Calculate percentage
print(f"\nThe model explains {R2_manual*100:.1f}% of the variance in data usage.")
```

```python
# Verify with predictions
Y_pred_manual = a + b1 * X1 + b2 * X2

# Also calculate using the traditional R² formula as verification
SS_total = np.sum((Y - np.mean(Y))**2)
SS_residual = np.sum((Y - Y_pred_manual)**2)
R2_traditional = 1 - (SS_residual / SS_total)

print(f"\nVerification:")
print(f"R² (your formula) = {R2_manual:.4f}")
print(f"R² (traditional formula) = {R2_traditional:.4f}")
```

```python
import numpy as np

# Raw data
X1 = phone_data['ST'].values
X2 = phone_data['MRC'].values
Y = phone_data['DU'].values
n = len(Y)

print("=" * 70)
print("CORRECT CALCULATION USING DEVIATION FORMULAS")
print("=" * 70)

# Step 1: Calculate raw sums (capital sigma notation)
SUM_X1 = np.sum(X1)
SUM_X2 = np.sum(X2)
SUM_Y = np.sum(Y)
SUM_X1_squared = np.sum(X1**2)
SUM_X2_squared = np.sum(X2**2)
SUM_Y_squared = np.sum(Y**2)
SUM_X1X2 = np.sum(X1 * X2)
SUM_X1Y = np.sum(X1 * Y)
SUM_X2Y = np.sum(X2 * Y)

print(f"\nRaw sums (Capital Σ):")
print(f"ΣX₁ = {SUM_X1:.1f}")
print(f"ΣX₂ = {SUM_X2:.0f}")
print(f"ΣY = {SUM_Y:.1f}")
print(f"ΣX₁² = {SUM_X1_squared:.2f}")
print(f"ΣX₂² = {SUM_X2_squared:.0f}")
print(f"ΣY² = {SUM_Y_squared:.2f}")
print(f"ΣX₁Y = {SUM_X1Y:.2f}")
print(f"ΣX₂Y = {SUM_X2Y:.1f}")
print(f"ΣX₁X₂ = {SUM_X1X2:.1f}")

# Step 2: Calculate deviation sums (lowercase sigma notation) - Image 2 formulas
sum_x1_squared = SUM_X1_squared - (SUM_X1**2 / n)
sum_x2_squared = SUM_X2_squared - (SUM_X2**2 / n)
sum_y_squared = SUM_Y_squared - (SUM_Y**2 / n)
sum_x1y = SUM_X1Y - (SUM_X1 * SUM_Y / n)
sum_x2y = SUM_X2Y - (SUM_X2 * SUM_Y / n)
sum_x1x2 = SUM_X1X2 - (SUM_X1 * SUM_X2 / n)

print(f"\nDeviation sums (lowercase Σ - from Image 2 formulas):")
print(f"Σx₁² = ΣX₁² - (ΣX₁)²/n = {sum_x1_squared:.4f}")
print(f"Σx₂² = ΣX₂² - (ΣX₂)²/n = {sum_x2_squared:.4f}")
print(f"Σy² = ΣY² - (ΣY)²/n = {sum_y_squared:.4f}")
print(f"Σx₁y = ΣX₁Y - (ΣX₁)(ΣY)/n = {sum_x1y:.4f}")
print(f"Σx₂y = ΣX₂Y - (ΣX₂)(ΣY)/n = {sum_x2y:.4f}")
print(f"Σx₁x₂ = ΣX₁X₂ - (ΣX₁)(ΣX₂)/n = {sum_x1x2:.4f}")

# Step 3: Calculate b1 and b2 using Image 1 formulas with DEVIATION sums
numerator_b1 = (sum_x2_squared * sum_x1y) - (sum_x2y * sum_x1x2)
denominator = (sum_x1_squared * sum_x2_squared) - (sum_x1x2**2)
b1 = numerator_b1 / denominator

numerator_b2 = (sum_x1_squared * sum_x2y) - (sum_x1y * sum_x1x2)
b2 = numerator_b2 / denominator

a = (SUM_Y - b1 * SUM_X1 - b2 * SUM_X2) / n

print(f"\nRegression Coefficients:")
print(f"b₁ = {b1:.6f}")
print(f"b₂ = {b2:.6f}")
print(f"a = {a:.4f}")
print(f"\nEquation: Y = {a:.2f} + {b1:.4f}X₁ + {b2:.5f}X₂")

# Step 4: Calculate R² using Image 3 formula with DEVIATION sums
R2_correct = (b1 * sum_x1y + b2 * sum_x2y) / sum_y_squared

print(f"\n" + "=" * 70)
print(f"R² = (b₁Σx₁y + b₂Σx₂y) / Σy²")
print(f"R² = ({b1:.6f} × {sum_x1y:.4f} + {b2:.6f} × {sum_x2y:.4f}) / {sum_y_squared:.4f}")
print(f"R² = {R2_correct:.4f}")
print(f"R² = {R2_correct:.2f} ← This should match your manual calculation!")
print("=" * 70)
```
